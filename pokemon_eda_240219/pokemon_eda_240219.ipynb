## 정규 표현식
- [A-Z] : A부터 Z까지의 대문자 중 한 가지로 시작하고,
- [a-z] : 그 뒤에 a부터 z까지의 소문자 중 한 가지가 붙는데,
- ___*___ : 그 소문자의 개수는 하나 이상인 패턴 (*는 정규표현식 중에서 "반복"을 나타내는 기호)


```python
import re

name = "CharizardMega Charizard X"
```

```python
name_split = name.split(" ")
name_split
```

```python
temp = name_split[0]
temp
```

```python
tokens = re.findall('[A-Z][a-z]*', temp)
tokens
```

```python
tokens = []
for part_name in name_split:
    a = re.findall('[A-Z][a-z]*', part_name)
    tokens.extend(a)
tokens
```

```python
def tokenize(name):
    tokens = re.findall('[A-Z][a-z]*', name)
    return np.array(tokens)
```

```python
name = "CharizardMega Charizard X"
tokenize(name)
```

```python
all_tokens = list(legendary['Name'].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)
    
print(len(set(token_set)))
print(token_set)
```

```python
from collections import Counter

a = [1,1,0,0,0,1,1,2,3]
Counter(a)
```

```python
Counter(a).most_common()
```

```python
most_common = Counter(token_set).most_common(10)
most_common
```

### 스케일

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
# X.columns
# X 

X_int = X.iloc[:,:9]
# X_int
X_obj = X.iloc[:,10:]
# X_obj
```

```python
X_int = mm.fit_transform(X_int)
# X_int
X_int = pd.DataFrame(X_int, columns=X.iloc[:,:9].columns)
X_int
```

```python
X_scale = pd.concat([X_int,X_obj], axis=1)
X_scale
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.2, random_state=205)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
model = DecisionTreeClassifier(random_state=205)
model
```

스케일을 하고 기본 학습을 진행했을 때 오히려 더 좋은 결과가 나오는 것을 확인
