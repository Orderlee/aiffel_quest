```python
import zipfile
import os

base_dir = os.getenv('HOME') + '/aiffel/aiffel_quest/rock_scissor_paper_240306/'

# 압축 파일 경로
zip_file = base_dir + "test2.zip"
# 압축 해제할 폴더 경로
extract_to_folder = base_dir + "data"

# data 폴더가 없다면 생성
os.makedirs(extract_to_folder, exist_ok=True)

# zipfile 모듈을 사용하여 압축 해제
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(extract_to_folder)

print("압축 해제가 완료되었습니다.")
```

```python
import os

def rename_files_in_folders(root_folder):
    # root_folder 내의 모든 폴더 탐색
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        # 디렉토리인 경우에만 처리
        if os.path.isdir(folder_path):
            # 폴더 내의 모든 파일 탐색 및 이름 변경
            for idx, file_name in enumerate(os.listdir(folder_path)):
                old_file_path = os.path.join(folder_path, file_name)
                new_file_name = f"{idx}_jy.jpg"
                new_file_path = os.path.join(folder_path, new_file_name)
                os.rename(old_file_path, new_file_path)
                print(f"파일 이름 변경: {old_file_path} -> {new_file_path}")

# 이미지가 있는 폴더의 경로
root_folder = base_dir +"/data/test"

# 각 폴더의 파일 이름 변경
rename_files_in_folders(root_folder)
```

```python
import os
import shutil

def move_files(source_folder, dest_folder):
    # 폴더가 없으면 생성
    os.makedirs(dest_folder, exist_ok=True)
    
    # 폴더 내의 모든 파일을 이동
    for file_name in os.listdir(source_folder):
        source_path = os.path.join(source_folder, file_name)
        dest_path = os.path.join(dest_folder, file_name)
        shutil.move(source_path, dest_path)
        print(f"{source_path} -> {dest_path} 이동 완료")

# 이미지가 있는 폴더의 경로
source_folder = base_dir+ 'data/test2/rock'
dest_folder = base_dir + 'data/rock'

# 파일 이동
move_files(source_folder, dest_folder)
```

```python
import shutil
import os

def compress_folder(folder_path):
    # 압축할 폴더의 경로와 이름
    folder_name = os.path.basename(folder_path)
    zip_file_path = folder_path + '.zip'

    # 폴더를 zip 파일로 압축
    shutil.make_archive(folder_path, 'zip', folder_path)

    print(f"'{folder_path}' 폴더를 압축하여 '{zip_file_path}' 파일로 저장했습니다.")

# 폴더의 경로
folder_path = base_dir + 'data/paper'

# 폴더를 압축
compress_folder(folder_path)
```
